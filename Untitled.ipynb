```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from tqdm import tqdm_notebook as tqdm
```

```python
aisles = pd.read_csv('aisles.csv')
departments = pd.read_csv('departments.csv')
prior = pd.read_csv('order_products__prior.csv')
train = pd.read_csv('order_products__train.csv')
order = pd.read_csv('orders.csv')
products = pd.read_csv('products.csv')
```

```python
df1 = pd.concat((train, prior), axis=0, ignore_index=True, sort=False)
```

```python
df1 = pd.merge(df1,order)
```

```python
df1['days_since_prior_order'] = df1['days_since_prior_order'].fillna(df1['days_since_prior_order'].mean())
```

```python
df1.head(5)
```

```python
days = df1.days_since_prior_order

days = days.groupby(days).transform('count')
df1['days'] = days/len(days)
```

```python
hour = df1.order_hour_of_day

hour = days.groupby(hour).transform('count')
df1['hour'] = hour/len(hour)
```

```python
dow  = df1.order_dow
dow = dow.groupby(dow).transform('count')
df1['dow'] = dow/len(dow)
```

```python
df1 = pd.merge(df1,products, on='product_id')
```

```python
df1['hours_of_week'] = df1['hour']*df1['dow']
```

```python
reordered = df1[df1["reordered"]==1]
```

```python
reordered
```

```python
items  = pd.merge(pd.merge(left=products, right=departments, how='left'), right=aisles, how='left')
reordered = pd.concat((reordered, items), axis=0, ignore_index=True, sort=False)
```

```python
reordered.head()
```

```python
grouped = reordered.groupby("department")["product_id"].aggregate({'Total_products': 'count'}).reset_index()
grouped['Ratio'] = grouped["Total_products"].apply(lambda x: x /grouped['Total_products'].sum())
grouped.sort_values(by='Total_products', ascending=False, inplace=True)
grouped
```

```python
products.head()
```

```python
grouped = reordered.groupby(["department", "aisle"])["product_id"].aggregate({'Total_products': 'count'}).reset_index()
grouped.sort_values(by='Total_products', ascending=False, inplace=True)
```

```python
grouped['Ratio'] = grouped["Total_products"].apply(lambda x: x /grouped['Total_products'].sum())
grouped.sort_values(by='Total_products', ascending=False, inplace=True)
grouped
```

```python
result = pd.concat((departments, grouped), axis=1, ignore_index=False, sort=False)
```

```python
final = pd.concat((df1, result), axis=1, ignore_index=False, sort=False)
```

```python
final.drop(['product_name','department','department','aisle'], axis=1, inplace=True)

```

```python
final
```

```python
prior = final[final.eval_set=='prior']
train = final[final.eval_set=='train']
```

```python
train.sort_index(axis=1, inplace=True)
prior.sort_index(axis=1,inplace=True)
```

```python
train
```

```python
y_train = prior.reordered
y_test = train.reordered
X_train = prior.drop(['reordered','eval_set'], axis=1, inplace=True)
X_test = train.drop(['reordered','eval_set'],axis=1,inplace=True)
```
